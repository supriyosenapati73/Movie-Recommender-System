```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
def remove_stopwords(text):
  new_text = []
  for i in text.split():
    if i not in stop_words:
      new_text.append(i)
  return " ".join(new_text)
```

```python
remove_stopwords(data['tags'][0])
```

```python
lemmatizer = WordNetLemmatizer()
def preprocess(text):
  text = str(text).lower()

  text = re.sub(r'[^a-zA-Z\s]','', text)

  words = text.split();
  words = [word for word in words if word not in stop_words]

  words = [lemmatizer.lemmatize(word) for word in words]

  return " ".join(words)
```

```python
preprocess(data['tags'][0])
```

```python
data['tags'] = data['tags'].apply(preprocess)
```

```python
data = data.reset_index(drop = True)
```

```python
indices = pd.Series(data.index, index = data['title']).drop_duplicates()
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features = 50000, ngram_range=(1,2))
```

```python
tfidf_matrix = tfidf.fit_transform(data['tags'])
```

```python
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
top = sorted(list(enumerate(cosine_similarity(tfidf_matrix[0],tfidf_matrix).ravel())),key = lambda x:x[1], reverse = True)[1:11]
```

```python
top[0][0]
```

```python
indices.index[2996]
```

```python
for i in top:
  movie = i[0]
  print(indices.index[movie])
```

```python
def recommend(movie):
  if movie not in indices.index:
    return "Movie not found"
  movie_index = indices[movie]
  top_10 = sorted(list(enumerate(cosine_similarity(tfidf_matrix[movie_index],tfidf_matrix).ravel())),key = lambda x:x[1], reverse = True)[1:11]
  for i in top_10:
    movies = i[0]
    print(indices.index[movies])
```

```python
recommend('Toy Story 2')
```
